```python
df.groupby('gender')['time_spent'].mean()
```

```python
import matplotlib.pyplot as plt
df.groupby('gender')['time_spent'].mean().plot(kind='bar',color=['pink','skyblue'])
plt.title('Average Time Spent by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Time Spent (hours)')
plt.show()
```

```python
df['age_group']=pd.cut(df['age'],bins=[0,18,25,35,50,100],labels=['0-18','19-25','26-35','36-50','51+'])
```

```python
df.groupby('gender')['time_spent'].mean().plot(kind='bar',color=['pink','skyblue'])
plt.title('Average Time Spent by age group')
plt.xlabel('age group')
plt.ylabel('Average Time Spent (hours)')
plt.show()
```

```python
df.groupby('age_group')['time_spent'].mean().plot(kind='bar',color=['orange'])
plt.title('Average Time Spent by age group')
plt.xlabel('age group')
plt.ylabel('Average Time Spent (hours)')
plt.show()
```
